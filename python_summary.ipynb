# 새로운

```python

# i와 j의 조합에 해당하는 값
i_values = ['Name', 'Find']
j_values = [0, 1,2, 3, 4, 5]

# 표를 만들기 위한 빈 DataFrame
df = pd.DataFrame(index=i_values, columns=j_values)

start_value = 10
for j in j_values:
    df.loc['Name', j] = start_value + j
    df.loc['Find', j] = start_value+ j


df
```
